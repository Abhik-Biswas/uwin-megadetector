```python
import os
import pandas as pd

# Define the folder containing the CSV files
folder_path = r"D:\UWIN Tagged Photos Outputs\yolo_formatted_csv"

# List all CSV files in the folder
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Initialize an empty list to store the DataFrames
dataframes = []

# Read each CSV file and append it to the list
for i, file in enumerate(csv_files):
    df = pd.read_csv(os.path.join(folder_path, file))
    dataframes.append(df)

# Concatenate all DataFrames into one
concatenated_df = pd.concat(dataframes, ignore_index=True)
```

```python
import os
import shutil
import pandas as pd
from PIL import Image
from tqdm import tqdm

def convert_bbox_to_yolo(bbox_md):
    """
    Convert bounding box from MegaDetector format to YOLOv5 format and ensure values are within [0, 1].

    Parameters:
    bbox_md (list): Bounding box in MegaDetector format [x_min, y_min, width_of_box, height_of_box]

    Returns:
    list: Bounding box in YOLOv5 format [c_x, c_y, width, height] with values clamped to [0, 1]
    """
    x_min, y_min, width_of_box, height_of_box = bbox_md
    
    # Calculate center coordinates
    c_x = x_min + width_of_box / 2
    c_y = y_min + height_of_box / 2
    
    # Ensure all values are within the range [0, 1]
    c_x = min(max(c_x, 0), 1)
    c_y = min(max(c_y, 0), 1)
    width_of_box = min(max(width_of_box, 0), 1)
    height_of_box = min(max(height_of_box, 0), 1)
    
    return [c_x, c_y, width_of_box, height_of_box]


def process_images_and_annotations(df, images_dir, labels_dir):
    if not os.path.exists(images_dir):
        os.makedirs(images_dir)
    if not os.path.exists(labels_dir):
        os.makedirs(labels_dir)
    
    for i in tqdm(range(df.shape[0])):
        row = df.iloc[i,:]
        image_path = row['photoAbsPath']
        image = Image.open(image_path)
        image_width, image_height = image.size
        
        # Copy the image to the images directory
        image_name = os.path.basename(image_path)
        shutil.copy(image_path, os.path.join(images_dir, image_name))
        
        # Convert bounding box to YOLO format
        bbox = [row['x'], row['y'], row['w'], row['h']]
        yolo_bbox = convert_bbox_to_yolo(bbox)
        
        # Create the annotation file
        label_file_path = os.path.join(labels_dir, image_name.replace('.jpg', '.txt').replace('.JPG', '.txt'))
        with open(label_file_path, 'w') as label_file:
            # Assuming there's a column 'category' for the class label
            category = row['category']
            label_file.write(f"{category} {' '.join(map(str, yolo_bbox))}\n")
```

```python
# Example usage:
images_dir = r"D:\dataset\images"
labels_dir = r"D:\dataset\labels"

# Assuming `concatenated_df` is already defined
process_images_and_annotations(concatenated_df, images_dir, labels_dir)
```

```python
def is_text_file(file_path):
    """
    Check if a file contains text data by attempting to read it as text.
    """
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            file.read()
        return True
    except:
        return False

def convert_jpg_to_txt(directory):
    """
    Convert .jpg files containing text data to .txt files in the specified directory.
    """
    for filename in tqdm(os.listdir(directory)):
        if filename.endswith('.TXT'):
            jpg_path = os.path.join(directory, filename)
            if is_text_file(jpg_path):
                txt_path = os.path.join(directory, filename.rsplit('.', 1)[0] + '.txt')
                os.rename(jpg_path, txt_path)
                print(f'Renamed: {jpg_path} to {txt_path}')
            else:
                print(f'Skipped: {jpg_path} (not a text file)')
```
